```python
import matplotlib.pyplot as plt
import numpy as np
import math
%matplotlib inline
```

```python
plt.plot([10,11,12],[3,5,9])
plt.show()
```

```python
x= np.arange(1,10)
y = x**2

x2 = np.arange(-1,15)
y2 = x2 + 50

plt.title("Squares")
plt.xlabel("x-values")
plt.ylabel("squared-values")
plt.xlim(-5,15)
plt.ylim(-20,120)
plt.plot(x,y,'r-',label='squares line')
plt.plot(x2,y2,label='linear line')
plt.legend()
plt.grid()
plt.text(7.5,60,'intersection')
plt.show()
```

```python
plt.plot(x,y,'b--',x2,y2,'ro')
plt.show()
```

```python
x = np.arange(-10,10)
sigmoid = lambda x: 1/(1+math.exp(-x))
#fx = 1/(1+np.vectorize(math.exp)(-x))
fx = [sigmoid(i) for i in x]
plt.plot(x,fx)
plt.show()
```

```python
x = np.arange(1,10,0.1)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.figure(figsize=(15,10))
plt.subplot(2,3,1)
plt.plot(x,y)

plt.subplot(2,3,5)
plt.plot(x,y2)

plt.subplot(2,3,3)
plt.plot(x,y3)

plt.show()
```
