```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
uniform_data = stats.uniform.rvs(loc=1,scale=100,size=100000)
```

```python
sns.histplot(uniform_data,bins=20,stat='density')
plt.show()
```

```python
sns.displot(uniform_data)
plt.show()
```

```python
stats.uniform.pdf(-10,loc=1,scale=100)
```

```python
stats.uniform.cdf(200,loc=1,scale=100)
```
